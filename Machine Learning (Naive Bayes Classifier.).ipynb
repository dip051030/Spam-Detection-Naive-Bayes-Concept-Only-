```python
import pandas as pd
df = pd.read_table(filepath_or_buffer=r'/home/dyane__4/Desktop/Datasets/SMSSpamCollection', names = ['label', 'SMS_message'])
```

```python
df.head()
```

```python
df['label'] = df['label'].map({'ham':0, 'spam':1})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['SMS_message'], df['label'], stratify=df['label'])
```

```python
X_test
```

```python
print('Number of rows in the total set: {}'.format(df.shape[0]))
print('Number of rows in the training set: {}'.format(X_train.shape[0]))
print('Number of rows in the test set: {}'.format(X_test.shape[0]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectoriser = CountVectorizer()

training_data = count_vectoriser.fit_transform(X_train)
testing_data = count_vectoriser.transform(X_test)
```

```python
print(training_data)
```

```python
from sklearn.naive_bayes import MultinomialNB

naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
print(testing_data)
```

```python
predictions = naive_bayes.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
print('Precision score: ', format(precision_score(y_test, predictions)))
print('Recall score: ', format(recall_score(y_test, predictions)))
print('F1 score: ', format(f1_score(y_test, predictions)))
```

```python
def classify_sms_interactive():
    """
    Interactive function to classify SMS messages as spam or ham.
    Allows users to input messages and get predictions until they choose to quit.
    """
    print("\nSMS Spam Classifier")
    print("------------------")
    print("Enter a message to classify it as spam or ham.")
    print("Type 'quit' or 'exit' to end the program.\n")

    while True:
        # Get user input
        user_message = input("Enter your message: ").strip()

        # Check for exit command
        if user_message.lower() in ['quit', 'exit', 'q']:
            print("\nThank you for using the SMS Classifier!")
            break

        # Handle empty input
        if not user_message:
            print("Please enter a message or type 'quit' to exit.")
            continue

        try:
            # Get prediction

            # Calculate prediction probability
            text_vectorized = count_vectoriser.transform([user_message])
            result = naive_bayes.predict(text_vectorized)
            probability = naive_bayes.predict_proba(text_vectorized)[0]

            # Format the output
            print("\nResult:")
            print(f'For text:  "{user_message}"')
            print(f"Classification: {['not spam' if result[0] == 0 else 'Spam'][0]}")
            print(f"Confidence: {max(probability) * 100:.2f}%")
            print("-" * 50 + "\n")

        except Exception as e:
            print(f"Error processing message: {str(e)}")
            print("Please try again with a different message.\n")


# Add this at the end of your notebook to start the interactive session
if __name__ == "__main__":
    classify_sms_interactive()
```
